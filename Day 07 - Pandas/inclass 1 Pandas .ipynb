# $$Pandas$$

- Pandas is a python library
- provides fast, flexible, and expressive data structures designed to make working with tables easy
- Tabular data with heterogeneously-typed columns, as in an SQL table or Excel spreadsheet
- [Documentation](https://pandas.pydata.org/docs/getting_started/overview.html)
- 
- Data Structures in pandas are:

Dimensions | Name | Description
----|---|---
1 | Series  | 1D   labeled homogeneously-typed array
2 | DataFrame | General 2D   labeled, size-mutable tabular structure with potentially heterogeneously-typed column

![](https://pandas.pydata.org/docs/_images/01_table_dataframe.svg)

- main purpose is commonly ---> data prepration / data pre-processing
- You can explore, clean, and transform data before analytics or building ML Models
- 
-  we can :
    - Read from csv files
    - select rows and columns
    - filter, sort, group data
    - handle missing values
    - prepare data for analytics or ML

##### Basic Exploration

## Basic data accessing in a dataframe
- access a single column --- using `[]` or `.` operator
    - `df['column_name']` ---> Return a series object
    - show the `name` and the `data type` as well
    - 
- access a value in a column
    - `df['column_name'][index_value]` ---> gets to a particular value

```python
temp_df['city'][5]
type(temp_df['city'][5])

type(temp_df['Temperature'][5]) # Nan values in dataframes are empty numpy values

temp_df.Temperature
```
