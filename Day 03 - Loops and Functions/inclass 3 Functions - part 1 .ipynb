### Function Scope
- global scope, function scope
- Global variables ---> 
    - can be *accessed anywhere* in the scipt
- Local Variables ---> 
    - defined *within the function scope*
    -  can only be accessed within a function

````python
scalar = 1000 # GLOBAL SCOPE

# take number of kilos ---> return how many grams are in them

def grams(kilos):
    result = kilos * scalar # FUNCTION SCOPE --- LOCAL SCOPE
    # local variable
    print("i can see what result is: ", result)
    return result


grams(4)
````
